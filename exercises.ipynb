```python
from functools import reduce
from random import randint

# Dataset generation function (from your screenshot)
def generate_random_student_dataset(num_students=50):
    names = [f"Student {i}" for i in range(1, num_students + 1)]
    dataset = [
        {"name": name, "scores": [randint(50, 100) for _ in range(randint(3, 6))]}
        for name in names
    ]
    return dataset

# Generate the random dataset
students = generate_random_student_dataset(50)

#  solution using map, reduce, and filter
students_with_avg = list(
    map(
        lambda student: {
            "name": student["name"],
            "average_score": reduce(lambda a, b: a + b, student["scores"]) / len(student["scores"]),
        },
        students,
    )
)

filtered_students = list(
    filter(lambda student: student["average_score"] > 90, students_with_avg)
)

# Print the result
print(filtered_students)
```

```python
from functools import reduce
from collections import defaultdict

# Example dataset
products = [
    {"name": "Product A", "price": 50, "category": "Electronics"},
    {"name": "Product B", "price": 50, "category": "Electronics"},
    {"name": "Product C", "price": 120, "category": "Home"},
    {"name": "Product D", "price": 80, "category": "Home"},
    {"name": "Product E", "price": 90, "category": "Sports"},
]

# Step 1: Group products by category
grouped_products = defaultdict(list)
for product in products:
    grouped_products[product["category"]].append(product["price"])

# Step 2: Calculate the average price for each category using reduce and map
average_prices = list(
    map(
        lambda category: {
            "category": category,
            "average_price": reduce(lambda x, y: x + y, grouped_products[category]) / len(grouped_products[category]),
        },
        grouped_products,
    )
)

# Step 3: Filter categories with an average price > 50
result = list(filter(lambda item: item["average_price"] > 50, average_prices))

# Output the result
print(result)
# hints: 1) Group products by category (you don't need to use map reduce for this part), then 2) use map reduce paradigm to
# calculate the average price for each category and filter categories with an average price > 50
```

# Biopython

Write the following five functions to analyze global alignments between two sequences using Biopython's `pairwise2` module:

1. **countMatches(s1, s2)**  
   This function takes two sequences (`s1`, `s2`) aligned using global alignment (pairwise2.globalxx) of the same length. It returns the number of positions where the elements of both sequences match.

2. **countMismatches(s1, s2)**  
   This function takes two sequences (`s1`, `s2`) aligned using global alignment of the same length. It returns the number of positions where the elements of the two sequences are different (i.e., they are not gaps, and the characters do not match).

3. **countGapOpens(s1, s2)**  
   This function takes two sequences (`s1`, `s2`) aligned using global alignment of the same length. It returns the number of gap openings in the alignment (a gap is opened when a '-' appears in the sequence).

4. **countGapExtensions(s1, s2)**  
   This function takes two sequences (`s1`, `s2`) aligned using global alignment of the same length. It returns the number of gap extensions (where '-' continues in the alignment after an initial gap is opened).

5. **getScore(s1, s2, matchScore, mismatchPenalty, gapOpenPenalty, gapExtensionPenalty)**  
   This function takes two sequences (`s1`, `s2`) aligned using global alignment and returns the alignment score based on the provided scoring scheme: `matchScore` for matches, `mismatchPenalty` for mismatches, `gapOpenPenalty` for opening a gap, and `gapExtensionPenalty` for extending a gap.

```python


# Function to count matches
def countMatches(s1, s2):
    """Counts the number of matches in the aligned sequences."""
    return sum(1 for a, b in zip(s1, s2) if a == b and a != '-')

# Function to count mismatches
def countMismatches(s1, s2):
    """Counts the number of mismatches in the aligned sequences."""
    return sum(1 for a, b in zip(s1, s2) if a != b and a != '-' and b != '-')

# Function to count gap openings
def countGapOpens(s1, s2):
    """Counts the number of gap openings in the aligned sequences."""
    count = 0
    in_gap = False
    for a, b in zip(s1, s2):
        if (a == '-' or b == '-') and not in_gap:
            count += 1
            in_gap = True
        elif a != '-' and b != '-':
            in_gap = False
    return count

# Function to count gap extensions
def countGapExtensions(s1, s2):
    """Counts the number of gap extensions in the aligned sequences."""
    count = 0
    in_gap = False
    for a, b in zip(s1, s2):
        if (a == '-' or b == '-') and in_gap:
            count += 1
        elif (a == '-' or b == '-') and not in_gap:
            in_gap = True
        elif a != '-' and b != '-':
            in_gap = False
    return count

# Function to calculate the alignment score
def getScore(s1, s2, matchScore, mismatchPenalty, gapOpenPenalty, gapExtensionPenalty):
    """Calculates the alignment score based on the provided scoring scheme."""
    score = 0
    in_gap = False
    for a, b in zip(s1, s2):
        if a == b and a != '-':
            score += matchScore
        elif a != b and a != '-' and b != '-':
            score += mismatchPenalty
        elif (a == '-' or b == '-') and not in_gap:
            score += gapOpenPenalty
            in_gap = True
        elif (a == '-' or b == '-') and in_gap:
            score += gapExtensionPenalty
        elif a != '-' and b != '-':
            in_gap = False
    return score

```

Align the sequences of the [Interleukin-12](https://en.wikipedia.org/wiki/Interleukin_12) chain A (denoted as `s1`) from the file [`IL12A.fasta`](https://qcbsciprolab2020.readthedocs.io/en/latest/file_samples/IL12A.fasta) and the Interleukin-12 chain B (denoted as `s2`) from the file [`IL12B.fasta`](https://qcbsciprolab2020.readthedocs.io/en/latest/file_samples/IL12B.fasta) and check the score as computed from pairwise2 and from your functions.

```python
from Bio import pairwise2
from Bio.SeqIO import parse

# Step 1: Read the sequences from the FASTA files
def read_sequence_from_fasta(file_path):
    with open(file_path, "r") as file:
        seq_record = next(parse(file, "fasta"))
        return str(seq_record.seq)

# Read sequences from IL12A.fasta and IL12B.fasta
s1 = read_sequence_from_fasta("IL12A.fasta")  # Interleukin-12 chain A
s2 = read_sequence_from_fasta("IL12B.fasta")  # Interleukin-12 chain B

# Step 2: Align the sequences using pairwise2 (Biopython)
alignments = pairwise2.align.globalxx(s1, s2)

# Step 3: Compute alignment score using pairwise2
pairwise2_score = alignments[0][2]  # The score is at index 2 in the tuple returned by alignments
print(f"Pairwise2 Alignment Score: {pairwise2_score}")

# Step 4: Implement a custom alignment function using dynamic programming (Needleman-Wunsch)
def custom_alignment_score(seq1, seq2):
    # Initialize scoring parameters
    match_score = 1
    mismatch_penalty = -1
    gap_penalty = -2

    # Initialize the scoring matrix (dynamic programming table)
    len1, len2 = len(seq1), len(seq2)
    dp = [[0] * (len2 + 1) for _ in range(len1 + 1)]

    # Fill the dynamic programming table
    for i in range(1, len1 + 1):
        for j in range(1, len2 + 1):
            match = dp[i - 1][j - 1] + (match_score if seq1[i - 1] == seq2[j - 1] else mismatch_penalty)
            delete = dp[i - 1][j] + gap_penalty
            insert = dp[i][j - 1] + gap_penalty
            dp[i][j] = max(match, delete, insert)

    # The final score is the value in the bottom-right corner of the matrix
    return dp[len1][len2]

# Step 5: Compute the custom alignment score
custom_score = custom_alignment_score(s1, s2)
print(f"Custom Alignment Score: {custom_score}")

#ouput 
# Pairwise2 Alignment Score: 103.0
# Custom Alignment Score: -166
```
